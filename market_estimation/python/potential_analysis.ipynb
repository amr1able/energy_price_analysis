```python
import os
```

```python
import pandas as pd
```

```python
data_in_dir = r"../data/"
for elt in os.listdir(data_in_dir): print(elt)
```

```python
energy_price_fname = r"Spotmarktpreis [2026-02-04 05-51-12].csv"
energy_price_pfname = data_in_dir + energy_price_fname
print(f"energy market data is in:  {energy_price_pfname:s}")
```

```python
# read the header of the file
header = open(energy_price_pfname, encoding="cp1252").readline().split(";")
# ";" is the delimiter so it was used in the .split : )
header
```

```python
df_energy_price = pd.read_csv(energy_price_pfname, sep = ";") 
#,index_col=["something"],parse_dates=["Datum"]
```

```python
df_energy_price
```

```python
df_energy_price.columns
```

```python
df_energy_price['Date_and_time'] = df_energy_price['Datum'] + ' ' + df_energy_price['von']
```

```python
df_energy_price
```

```python
df_energy_price["Date_and_time"] = pd.to_datetime(
    df_energy_price["Date_and_time"],
    format="%d.%m.%Y %H:%M",
    errors="raise"
)
```

```python
df_energy_price.dtypes
```

```python
df_energy_price.set_index("Date_and_time", inplace=True)
```

```python
df_energy_price["Spot_market_price_in_€/MWh"] = df_energy_price["Spotmarktpreis in ct/kWh"] * 10
```

```python
df_energy_price["Spot_market_price_in_€/MWh"] = df_energy_price["Spot_market_price_in_€/MWh"].round(3) 
#3 decimal places
```

```python
df_energy_price.columns
```

```python
df_energy_price.drop(columns=['Datum', 'von', 'Zeitzone von', 'bis', 'Zeitzone bis', 'Spotmarktpreis in ct/kWh'], inplace=True)
```

```python
df_energy_price
```

```python
file_CSV_name = "energy_price_clean.csv"
```

```python
df_energy_pfname = data_in_dir + file_CSV_name
df_energy_pfname
```

```python
df_energy_price.to_csv(df_energy_pfname,sep =";") # fancy extra thing to save the file as a CSV for absolutely no reason
```

Analysis begins

```python
df_energy_price["delta"] = df_energy_price["Spot_market_price_in_€/MWh"].diff()
df_energy_price
```

Calculating only Upward Variation

```python
df_energy_price["positive_delta"] = df_energy_price["delta"].clip(lower=0)
df_energy_price
```

```python
daily_basis = df_energy_price.groupby(df_energy_price.index.date).agg(
    min_price=("Spot_market_price_in_€/MWh", "min"),
    max_price=("Spot_market_price_in_€/MWh", "max"),
    mean_price=("Spot_market_price_in_€/MWh", "mean"),
    std_price=("Spot_market_price_in_€/MWh", "std"),
    upward_price_variation=("positive_delta", "sum"),
)
daily_basis["spread"] = daily_basis["max_price"] - daily_basis["min_price"]
```

```python
daily_basis
```

```python
daily_basis.columns
```

```python
for x in daily_basis.columns:
    daily_basis[x] = daily_basis[x].round(3) 
    #3 decimal places
```

```python
daily_basis
```

```python
file_daily_CSV_name = "energy_price_daily.csv"
```

```python
df_energy_daily_pfname = data_in_dir + file_daily_CSV_name
df_energy_daily_pfname
```

```python
daily_basis.to_csv(df_energy_daily_pfname,sep =";") # fancy extra thing to save the file as a CSV for absolutely no reason
```

Monthly Analysis

```python
daily_basis.index = pd.to_datetime(daily_basis.index)
```

```python
monthly_basis = daily_basis.groupby(daily_basis.index.to_period("M")).agg(
    min_daily_upward_price_variation=("upward_price_variation", "min"),
    max_daily_upward_price_variation=("upward_price_variation", "max"),
    mean_daily_upward_price_variation=("upward_price_variation", "mean"),
    std_daily_upward_price_variation=("upward_price_variation", "std"),
    total_sum_upward_price_variation_for_month=("upward_price_variation", "sum"),
)
monthly_basis["monthly_price_variation_spread"] = monthly_basis["max_daily_upward_price_variation"] - monthly_basis["min_daily_upward_price_variation"]
```

```python
monthly_basis
```

```python
for x in monthly_basis.columns:
    monthly_basis[x] = monthly_basis[x].round(3) 
    #3 decimal places
```

```python
file_monthly_CSV_name = "energy_price_potential_monthly.csv"
```

```python
df_energy_monthly_pfname = data_in_dir + file_monthly_CSV_name
df_energy_monthly_pfname
```

```python
monthly_basis.to_csv(df_energy_monthly_pfname,sep =";") # fancy extra thing to save the file as a CSV for absolutely no reason
```

Yearly Analysis

```python
yearly_basis = daily_basis.groupby(daily_basis.index.year).agg(
    min_daily_upward_price_variation=("upward_price_variation", "min"),
    max_daily_upward_price_variation=("upward_price_variation", "max"),
    mean_daily_upward_price_variation=("upward_price_variation", "mean"),
    std_daily_upward_price_variation=("upward_price_variation", "std"),
    total_sum_upward_price_variation_for_year=("upward_price_variation", "sum"),
)
yearly_basis["monthly_price_variation_spread"] = yearly_basis["max_daily_upward_price_variation"] - yearly_basis["min_daily_upward_price_variation"]
```

```python
yearly_basis
```

```python
for x in yearly_basis.columns:
    yearly_basis[x] = yearly_basis[x].round(3) 
    #3 decimal places
```

```python
file_yearly_CSV_name = "energy_price_potential_yearly.csv"
```

```python
df_energy_yearly_pfname = data_in_dir + file_yearly_CSV_name
df_energy_yearly_pfname
```

```python
yearly_basis.to_csv(df_energy_yearly_pfname,sep =";") # fancy extra thing to save the file as a CSV for absolutely no reason
```
